>##### E. Calculate IQR(the difference between 75% and 25% quartile) for `Hourly_Income` and `Hourly_Expense`.

```python
# Calculating Inter Quartile Range for 'Hourly_Income'

q1_hi= df1['Hourly_Income'].quantile(0.25)
q3_hi= df1['Hourly_Income'].quantile(0.75)
iqr_hi= q3_hi - q1_hi

print("Inter Quartile Range for 'Hourly_Income':", iqr_hi)

# Calculating Inter Quartile Range for 'Hourly_Expense'

q1_he= df1['Hourly_Expense'].quantile(0.25)
q3_he= df1['Hourly_Expense'].quantile(0.75)
iqr_he= q3_he - q1_he

print("Inter Quartile Range for 'Hourly_Expense':", iqr_he)
```

>#### A. Write a program
    Step 1: Create a  condition = 'LoanStatus'== 'Completed'| 'LoanStatus' == 'Defaulted' |\'LoanStatus' == 'Chargedoff'
    Step2: Create a user define function  using condition and `LoanStatus` column.

```python
def get_loan_status_count(data):
    try:
        if(data=='Completed'):
            return "Completed"
        if(data=="Defaulted"):
            return "Defaulted"
        if(data=="Chargedoff"):
            return "Defaulted"

        return " "  
    except Exception as e:
        pass
    
        
        
df2["LoanStatus"]=df2["LoanStatus"].apply(get_loan_status_count)

df2[df2['LoanStatus']!=" "]['LoanStatus'].value_counts()
```

>#### B. Write a program
    Step 1: Create a  dictionary called categories = 1: 'Debt Consolidation', 2: 'Home Improvement', 3: 'Business', 6: 'Auto', 7: 'Other'
    Step2: Create a user define function  using categories and `ListingCategory (numeric)` column.

```python
categories= {1: 'Debt Consolidation', 2: 'Home Improvement', 3: 'Business', 6: 'Auto', 7: 'Other'}

def get_categories(data):
    try:
        category=categories.get(data)
        return category
    except Exception as e:
        print(e)
   


 
df2['ListingCategory (numeric)'] = df2['ListingCategory (numeric)'].apply(get_categories)

df2['ListingCategory (numeric)'].value_counts()
```
